```python
%load_ext autoreload
%autoreload 2
import sys
import os
from pathlib import Path
import importlib
import random

%pip install -q -e  ../. 

from directory_issues import Source
import mediacloud.api

token = "BYO"
mc_dir = mediacloud.api.DirectoryApi(token)
```

```python
#Here, just grab a sample of sources, just check it out. 
sources_with_url_search_strings = []

more_sources = True
offset = 0
limit=100000
while more_sources and len(sources_with_url_search_strings) < 200:
    
    resp = mc_dir.source_list(limit = limit, offset=offset)
    matches = [r for r in resp["results"] if r["url_search_string"] is not None]
    S = [Source(r) for r in matches]
    for s in S:
        print(s)
        s.find_issues()
    sources_with_url_search_strings.extend(S)
    print(len(sources_with_url_search_strings))
    if resp["next"] is not None:
        offset += limit
    else:
        more_sources = False
```
